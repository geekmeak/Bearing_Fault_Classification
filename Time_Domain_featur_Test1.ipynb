```python
import numpy as np
import pandas as pd
```

```python
test1 = "C:/Users/ashees/Desktop/IMS/1st_test"
```

```python
import os
import glob
import itertools
```

```python
test1_file = glob.glob(test1 + "/*")
print("Total number of files inn test1: ", len(test1_file))
```

```python
T1 = []
for i in range(len(test1_file)):
    
    temp = pd.read_csv(test1_file[i], sep='\s+', header=None)
    T1.append(temp)
```

# Segmentation

## Mean

```python
mean_ch1 = []
mean_ch2 = []
mean_ch3 = []
mean_ch4 = []
mean_ch5 = []
mean_ch6 = []
mean_ch7 = []
mean_ch8 = []

for i in range(len(T1)):
    samp = T1[i].mean()
    mean_ch1.append(samp[0])
    mean_ch2.append(samp[1])
    mean_ch3.append(samp[2])
    mean_ch4.append(samp[3])
    mean_ch5.append(samp[4])
    mean_ch6.append(samp[5])
    mean_ch7.append(samp[6])
    mean_ch8.append(samp[7])
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30,20))


plt.subplot(421)
plt.plot(mean_ch1,'g')
plt.title('Bearing_1 ch1')

plt.subplot(422)
plt.plot(mean_ch2,'g')
plt.title('Bearing_1 ch2')

plt.subplot(423)
plt.plot(mean_ch3,'b')
plt.title('Bearing_2 ch1')

plt.subplot(424)
plt.plot(mean_ch4,'b')
plt.title('Bearing_2 ch4')

plt.subplot(425)
plt.plot(mean_ch5,'m')
plt.title('Bearing_3 ch5')

plt.subplot(426)
plt.plot(mean_ch6,'m')
plt.title('Bearing_3 ch6')

plt.subplot(427)
plt.plot(mean_ch7,'r')
plt.title('Bearing_4 ch7')

plt.subplot(428)
plt.plot(mean_ch8,'r')
plt.title('Bearing_4 ch8')

plt.show()
```

## RMS

```python
rms_ch1 = []
rms_ch2 = []
rms_ch3 = []
rms_ch4 = []
rms_ch5 = []
rms_ch6 = []
rms_ch7 = []
rms_ch8 = []

for i in range(len(T1)):
    rms = np.sqrt((T1[i] **2).mean())

    rms_ch1.append(rms[0])
    rms_ch2.append(rms[1])
    rms_ch3.append(rms[2])
    rms_ch4.append(rms[3])
    rms_ch5.append(rms[4])
    rms_ch6.append(rms[5])
    rms_ch7.append(rms[6])
    rms_ch8.append(rms[7])
```

```python
plt.figure(figsize=(30,20))


plt.subplot(421)
plt.plot(rms_ch1,'g')
plt.title('Bearing_1 ch1')

plt.subplot(422)
plt.plot(rms_ch2,'g')
plt.title('Bearing_1 ch2')

plt.subplot(423)
plt.plot(rms_ch3,'b')
plt.title('Bearing_2 ch1')

plt.subplot(424)
plt.plot(rms_ch4,'b')
plt.title('Bearing_2 ch4')

plt.subplot(425)
plt.plot(rms_ch5,'m')
plt.title('Bearing_3 ch5')

plt.subplot(426)
plt.plot(rms_ch6,'m')
plt.title('Bearing_3 ch6')

plt.subplot(427)
plt.plot(rms_ch7,'r')
plt.title('Bearing_4 ch7')

plt.subplot(428)
plt.plot(rms_ch8,'r')
plt.title('Bearing_4 ch8')

plt.show()
```
